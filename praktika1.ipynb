```python
from random import randint
speed = 0.3

w1 = [randint(1, 100) / 100, randint(1, 100) / 100]
w2 = [randint(1, 100) / 100, randint(1, 100) / 100]
w3 = [randint(1, 100) / 100, randint(1, 100) / 100]
x_train = [[1, 1], [1, 0], [0, 1], [0, 0]]
y_train1 = [0, 0, 1, 0]
y_train2 = [0, 1, 0, 0]
y_train3 = [0, 1, 1, 0]
```

```python
def predict(x_train, w):
  summator = 0
  for i in range(len(x_train)):
    summator += x_train[i] * w[i]
  return 1 if summator >= 1 else 0
```

```python
def train(x_train, y_train, w):
  for i in range(len(x_train)):
    error = y_train[i] - predict(x_train[i], w)
    if error != 0:
      for j in range(len(w)):
        w[j] += error * speed * x_train[i][j]
    print(w, error, y_train[i])
  print("\n")
```

```python
def epoch(count, train, x_train,y_train1, y_train2, y_train3, w1, w2, w3):
  print(f"Параметры весов:\n")
  for i in range(count):
    train(x_train, y_train1, w1)
    train(x_train, y_train2, w2)
    
    x_prom = []
    for i in range(len(x_train)):
      x_prom.append([])
      x_prom[i].append(predict(x_train[i], w1))
      x_prom[i].append(predict(x_train[i], w2))
    
    train(x_prom, y_train3, w3)
    print("-------------------------------------------")
```

```python
epoch(10, train, x_train, y_train1, y_train2, y_train3, w1, w2, w3)
print("\nКонечные веса: ",w1, "\n", w2, "\n", w3)
```

```python
print(predict([predict([1, 1], w1), predict([1, 1], w2)], w3))
print(predict([predict([1, 0], w1), predict([1, 0], w2)], w3))
print(predict([predict([0, 1], w1), predict([0, 1], w2)], w3))
print(predict([predict([0, 0], w1), predict([0, 0], w2)], w3))
```
